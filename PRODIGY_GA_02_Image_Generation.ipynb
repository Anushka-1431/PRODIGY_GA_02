```python
import diffusers
import torch

print("Libraries Installed Successfully")
print("Diffusers Version:", diffusers.__version__)
```

```python
from diffusers import StableDiffusionPipeline
import torch

pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")

print("Stable Diffusion Model Loaded Successfully")
```

```python
prompts = [
    "A futuristic AI city at night with glowing skyscrapers",
    "A humanoid robot teaching students in a smart classroom",
    "An AI-powered hospital in the year 2050",
    "A robot painting digital artwork in a modern studio",
    "A futuristic research lab with holographic displays"
]

for i, prompt in enumerate(prompts):
    print(f"\nGenerating Image {i+1}...")

    image = pipe(prompt).images[0]

    image.save(f"image_{i+1}.png")

    print(f"Saved image_{i+1}.png")
```
